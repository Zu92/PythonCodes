```python
board= ["" for x in range(10)]
board[0]="notboard"
freeSpaces=[1,2,3,4,5,6,7,8,9]

def resetGame(board,freeSpaces):
    board= ["" for x in range(10)]
    board[0]="notboard"
    freeSpaces=[1,2,3,4,5,6,7,8,9]

def printBoard(board):
    print('')
    print('  ' + board[1] + ' | ' + board[2] + '  | ' + board[3])

    print('------------')
  
    print('  ' + board[4] + ' | ' + board[5] + '  | ' + board[6])
   
    print('------------')
 
    print('  ' + board[7] + ' | ' + board[8] + '  | ' + board[9])
    
    print('')
    print('')
    print('------------')
    
def possibleMove():
    for x in freeSpaces:
        if board[x]!="":
            freeSpaces.remove(x)
        
def move(space,turn):
    board[space]=turn
    possibleMove()
    

def winner(space,l):
    b=[x for x in board]
    b[space]=l
    return ((b[1] == l and b[2] == l and b[3] == l) or
    (b[4] == l and b[5] == l and b[6] == l) or
    (b[7] == l and b[8] == l and b[9] == l) or
    (b[1] == l and b[4] == l and b[7] == l) or
    (b[2] == l and b[5] == l and b[8] == l) or
    (b[3] == l and b[6] == l and b[9] == l) or
    (b[1] == l and b[5] == l and b[9] == l) or
    (b[3] == l and b[5] == l and b[7] == l))
  
def aiMove():
    turn="O"
    cornersOpen = []
    edgesOpen = []
    for space in freeSpaces:
        if winner(space,turn):
            move(space,turn)
            print("Computer WIN")
            printBoard(board)
            return False
    for space in freeSpaces:    
        if winner(space,"X"):
            move(space,turn)
            return True
    for space in freeSpaces:
        if space==5:
            move(5,turn)
            return True
        if space in [1,3,7,9]:
            cornersOpen.append(space)
        if len(cornersOpen) > 0:
            move(selectRandom(cornersOpen),turn)
            return True
        if space in [2,4,6,8]:
            edgesOpen.append(space)
        if len(edgesOpen) > 0:
            move(selectRandom(edgesOpen),turn)
            return True
    if freeSpaces==[]:
        print("Tie")
        printBoard(board)
        return False
    
def selectRandom(li):
    import random
    ln = len(li)
    r = random.randrange(0,ln)
    return li[r]

def game():
    game= True
    while game==True:
        printBoard(board)
        playerMove=int(input("Please enter a number for 1 to 9"))
        while (playerMove in freeSpaces) ==False: 
            print("This Space is not empty or is not possible")
            print(freeSpaces)
            playerMove=int(input("Please enter a number like the ones adove"))
        if winner(playerMove,"X"):
            move(playerMove,"X")
            print("You Win")
            printBoard(board)
            game= False
        move(playerMove,"X")
        printBoard(board)
        game=aiMove()
        if freeSpaces==[]:
            print("Tie")
            printBoard(board)
            game= False

name = input("Enter your name")
print("Welcome" , name )
print("-------------------")
print("try to win")
game()
print("Do you want to play again?")
answer=input("Enter Y/N")
while answer in ("Y","y"):
    print("Lets play again" , name )
    resetGame(board,freeSpaces)
    game()
    print("Do you want to play again?")
    answer=input("Enter Y/N")
print("Goodbye", name)
    
           
```
